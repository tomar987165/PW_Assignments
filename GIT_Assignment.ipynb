  1. Create a Git working directory with feature1.txt and feature2.txt in the master branch

  2. Create 3 branches develop, feature1 and feature2

  3. In develop branch create develop.txt, do not stage or commit it

  4. Stash this file and check out to feature1 branch

  5. Create new.txt file in feature1 branch, stage and commit this file

  6. Checkout to develop, unstash this file and commit

  7. Please submit all the Git commands used to do the above step>

  ''' Bash

    mkdir git-task                # Create a new folder named 'git-task'
    cd git-task                   # Move into the directory

    git init                      # Initialize an empty Git repository

    touch feature1.txt            # Create file feature1.txt
    touch feature2.txt            # Create file feature2.txt
    git add feature1.txt feature2.txt   # Stage both files
    git commit -m "Add feature1.txt and feature2.txt"  # Commit them to master branch

    git branch develop            # Create a new branch named develop
    git branch feature1           # Create a new branch named feature1
    git branch feature2           # Create a new branch named feature2

    git checkout develop          # Switch to develop branch
    touch develop.txt             # Create develop.txt but do NOT stage or commit it

    git stash                     # Stash the uncommitted changes (develop.txt)
    git checkout feature1         # Switch to feature1 branch

    touch new.txt                 # Create new.txt in feature1 branch
    git add new.txt               # Stage new.txt
    git commit -m "Add new.txt in feature1 branch"   # Commit new.txt

    git checkout develop          # Switch back to develop branch
    git stash pop                 # Apply the stashed file (develop.txt) and remove it from stash
    git add develop.txt           # Stage develop.txt
    git commit -m "Add develop.txt after unstash"    # Commit the file
