1. What is Linux?

  - Linux is a free open source operating system.

2. What is the difference between Hard Link & Soft Link?

  - Hard links are multiple directory entries that point directly to the  same file data, while soft links are separate files that contain a pointer to another file's name or path.

3. What is a Kernel in Linux?

  - The Linux kernel is the core component of the Linux operating system, acting as the bridge between software and hardware.

4. How do you create a user account?

  - we can create a new user by using 'adduser' command and setting the password.
    sudo adduser TestUser
    sudo passwd TestUser
  
5. What is the ‘grep’ command used for in Linux?

  - 'grep' Command used for searching some plain text data in single or multiple files.

8. You suspect that a particular process is consuming excessive CPU resources on your Linux server. How
would you identify and terminate this process?

  - I will run <top> command to identify the <PID> of the process consuming maximum CPU with the help of "E" toggle key to convert thos numbers into MB,GB,TB. Once identified, I try terminating it gracefully using kill <PID>. If it doesnt stop, I use kill -9 <PID>. Then I verify using ps or top that the process is terminated.”
